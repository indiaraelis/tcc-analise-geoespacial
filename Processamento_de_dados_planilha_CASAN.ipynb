```python
import pandas as pd

# Ler o arquivo Excel (após fazer o upload)
df = pd.read_excel('Teste 2 Python.xlsx')

# Passo 1: Normalizar e limpar os motivos
df['Motivo'] = df['Motivo'].str.lower().str.strip()  # Converter para minúsculas e remover espaços

# Definir as palavras que queremos agrupar
palavras_grupos = [
    'demora', 'fatura', 'deslocamento', 'cavalete',
    'agilidade', 'execução', 'leitura', 'prazo',
    'faturas', 'vazam', 'atraso'
]

# Passo 2: Criar uma nova coluna para os grupos como listas vazias
df['Grupo'] = [[] for _ in range(len(df))]  # Inicializa uma lista vazia para cada linha

# Passo 3: Agrupar os motivos de acordo com as palavras selecionadas
for palavra in palavras_grupos:
    # Atualiza a coluna 'Grupo' para adicionar a palavra se encontrada no motivo
    df.loc[df['Motivo'].str.contains(palavra, case=False), 'Grupo'] = df['Grupo'] + [[palavra]]

# Passo 4: Exibir a contagem de manifestações por grupo
contagem_grupos = df.explode('Grupo')['Grupo'].value_counts()

print("Contagem de manifestações por grupo:")
print(contagem_grupos)

# Passo 5: Exibir os motivos agrupados
motivos_agrupados = df[df['Grupo'].map(len) > 0][['Motivo', 'Grupo']]
print("\nMotivos agrupados:")
print(motivos_agrupados)

# Passo 6: Exportar os motivos agrupados para um arquivo Excel
output_file_path = 'motivos_agrupados.xlsx'  # Defina o nome do arquivo de saída
motivos_agrupados.to_excel(output_file_path, index=False)

print(f'\nMotivos agrupados salvos em: {output_file_path}')
```

```python
import pandas as pd

# Carregar o DataFrame a partir do arquivo Excel
file_path = '/Teste 2 Python.xlsx'
df = pd.read_excel(file_path)

# Exibir informações do DataFrame para verificar se todas as linhas foram carregadas corretamente
print(f"Número de linhas no DataFrame original: {len(df)}")
print(df.head())  # Exibir as primeiras linhas para verificar o conteúdo

# Palavras e grupos (união de 'fatura' e 'faturas' em 'Grupo 2')
palavras_grupos = {
    'demora': 'Grupo 1',
    'fatura': 'Grupo 2',  # 'faturas' será removido, pois pertence ao mesmo grupo
    'deslocamento': 'Grupo 3',
    'cavalete': 'Grupo 4',
    'agilidade': 'Grupo 5',
    'execução': 'Grupo 6',
    'leitura': 'Grupo 7',
    'prazo': 'Grupo 8',
    'vazam': 'Grupo 9',
    'atraso': 'Grupo 10',
}

# Criar colunas para cada grupo
for grupo in set(palavras_grupos.values()):
    df[grupo] = 0  # Inicializa a coluna com 0

# Verificar cada palavra em cada motivo e atualizar as colunas dos grupos
for palavra, grupo in palavras_grupos.items():
    mask = df['Motivo'].str.contains(palavra, case=False, na=False)  # Certifique-se de lidar com NaNs
    df.loc[mask, grupo] = 1  # Atualiza para 1 se o motivo pertence ao grupo

# Renomear as colunas para incluir o nome e o número do grupo
df.columns = [col if col not in palavras_grupos.values() else f"{col} - {col.split(' ')[1]}" for col in df.columns]

# Confirmar que o DataFrame foi atualizado
print(f"Número de linhas no DataFrame atualizado: {len(df)}")

# Exibir as primeiras e últimas linhas do DataFrame para confirmação
print(df.head(10))  # Primeiras 10 linhas
print(df.tail(10))  # Últimas 10 linhas

# Exportar o DataFrame completo para CSV
output_file_path = '/motivos_com_grupos_separados.csv'
df.to_csv(output_file_path, index=False)

# Usando Google Colab para download
from google.colab import files
files.download(output_file_path)
```
